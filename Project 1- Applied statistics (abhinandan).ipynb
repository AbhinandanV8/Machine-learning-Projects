### Question 2: 
An electrical manufacturing company conducts quality checks at specified periods on the
products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random
sample of 10 manufactured items is selected. Answer the following questions.
<br>A. Probability that none of the items are defective?<br>
B. Probability that exactly one of the items is defective?<br>
C. Probability that two or fewer of the items are defective?<br>
D. Probability that three or more of the items are defective ?

Solution:<br>

To solve this problem, we'll use Binomial distribution.
    here,sample n = 10<br>
        p = 0.05 or 5%
        k = array of 10 set (from 0 to 11)

```python
n=10
p=0.05
k=np.arange(0,11)
print(n,p,k)
```

```python
binomial = stats.binom.pmf(k,n,p)
binomial
```

```python
# sol A:
print('A. Probablity that --NO item is defective--',binomial[0])
```

```python
#sol B:
print('B. Probablity that --only one item is defective--',binomial[1])
```

```python
#sol C: can also be done with cumm. function
print('C. Probablity that --2 or fewer are defective--',
          (binomial[0]+binomial[1]+binomial[2]))
```

```python
cumm_binomial = stats.binom.cdf(k,n,p)
print(cumm_binomial)
```

```python
#sol D:
print('D. Probablity that --3 or more are defective--',1-cumm_binomial[2])
```

### Question 3:

### Question 5:
A group of 300 professionals sat for a competitive exam. The results show the information of
marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a
normal distribution. Answer the following questions.<br>
A. What is the percentage of students who score more than 80.<br>
B. What is the percentage of students who score less than 50.<br>
C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?<br>

Solution:

we will solve this question through the help of z-test.
test stat z = value - mean / SD 

```python
a=stats.norm.cdf(80,loc=60,scale=12)
print('% of Students scored less than 80=',a)

#sol A:
print('\n A. Percentage of students who score more than 80 =',(1-a)*100,'%')
```

```python
# Sol B:
b=stats.norm.cdf(50,60,12)
print('B. percentage of students who score less than 50 = ', b*100,'%')
```

#sol C:

We will have to use the Standard Normal Distribution table in reverse to solve this problem .<br>
We know that we have to find the top 10% of the class. i.e.<br> P(z=(x-60)/12))=0.1

Now, we know that top 10% means 0.9. hence, we search for 0.9(or close) value in table.<br>
In table 0.8997 is present. So, we will use the x-axis and y-axis values to calculate the value of 'Z'<br>
#### Z = 1.2(y) + 0.08(x) = 1.28

Now, we can work to find the X value from the above formula:<br>
x-60 / 12 = 1.28<br>
x-60 = 1.28 * 12<br>
x = 15.36 + 60 = 75.36<br>

#### The distinction marks should be 75.36

```python
# as a cross-check

c = stats.norm.cdf(75.36,60,12)
print('C. Percentage of students below marks 75.36 are',c*100)
print('or we can say almost 90%')
print(':)')
```

#### Hyposthesis 1:
 - H0: There is no significant difference between the Mean Funds raised by Operativg vs Closed companies.
 - H1: There is significant difference between the Mean Funds raised by Operating vs Closed companies.
 
 To check for the significane, we'll use T-Test.<br>
    T-test: <br>
A t-test is a type of inferential statistic which is used to determine if there is a signifi cant differencebetween the means of two groups which may be related in certain features.

```python
df3.head(5)
```

```python
df3.groupby('OperatingState').mean()
```

```python
F_operatingC = np.array(df3[df3['OperatingState']=='Operating']['Funding'])
F_closedC = np.array(df3[df3['OperatingState']=='Closed']['Funding'])

print('Mean value of Funds from Operating company',format(F_operatingC.mean().round(2)))
print('Mean value of Funds from Closed company',format(F_closedC.mean().round(2)))
```

```python
#performing t-test
t, p_value = stats.ttest_ind(F_operatingC, F_closedC, axis = 0)

if p_value < 0.05:
    print(f'With the p-value as {p_value}, the difference is significant. Hence we reject the null Hypothesis.')
else:
    print(f'With the p-value as {p_value}, the difference is not significant.\nHence we fail to reject the null Hypothesis.')
        
```
